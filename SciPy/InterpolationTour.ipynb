## Finding Values with Boolean Masks

```python
theta = np.arange(0.,2*np.pi,np.pi/100)
plt.plot(theta, np.sin(theta))
plt.show()
```

Now suppose that we want to plot just the data with 0<sin(θ)<0.5. We’re going to have to "find” the data with 0<sin(θ) and sin(θ)<0.5.

```python
ii = (np.sin(theta)>0) & (np.sin(theta)<0.5)
plt.plot(theta,np.sin(theta),theta[ii],np.sin(theta[ii]),"g-")
plt.show()
```

We want to get rid of that horizontal line. To do this, we must set one value along the line to NaN. In this case, that element is where "y equals max(y)", so then set that element to NaN.

```python
y = np.sin(theta[ii])
jj = (y == max(y))
y[jj] = NaN
plt.plot(theta, np.sin(theta))
plt.plot(theta[ii],y,linewidth=4)
plt.show()
```
