# Partical Tracking over multiple pictures & Data Analysis

```python
from PIL import Image
import requests
from StringIO import StringIO
import Blobfinder as BF
import numpy as np
%pylab inline
```

```python
def TrackParticals(P, tau, deltaP):
    """
    TrackParticles takes P, the minimum mass of blobs, tau, the brightness threshold, and deltaP, 
    the minimum distance a blob must travel to be accounted for. This function runs through
    every picture in my run4 file using a for 3 for loops (one for each numbers place). Each itteration
    of the for loop does the following: First, two consecutive images are opened. Then, BlobFinder is run
    on both pictures, using tau as the brightness threshold to find the blobs in the picture. The 
    resulting list of blobs is stored in LstOfBlobs for one pic and LstOfBlobs2 for the other.
    Then, both lists are run through the getbeads function that picks out the blobs that have a mass
    attribute greater than P pixels. The resulting blobs are now reffered to as "beads" and are appended
    to LstOfBeads and LstOfBeads2. Next, we check each bead and find the distance from its center of
    mass to the center of mass of every bead in the next picture. I have it set up so that only distances
    less than deltaP pixels are appended to the temporary list 'temp' (this is so we aren't considering 
    the distance from one bead to another all the way across the picture). Finally, the minimum value 
    from the list 'temp' is added to 'distancesLst'. The function then returns a text file containing
    'distancesLst' with values to 4 digits of accuracy.
    """
    distancesLst = []
    #run4/frame00000 through run4/frame00009
    for i in range(0, 10):
        #if loop prevents picture name from exceeding expected values.
        if i < 9:
            pic1 = "run4/frame0000"+ str(i) + ".jpg"
            pic2 = "run4/frame0000"+ str(i+1) + ".jpg"
            #open two successive pictures
            im1 = Image.open(pic1)
            im2 = Image.open(pic2)
            
            #run BlobFinder to get a list of all bright objects in picture that includes
            #the mass and center of mass.
            LstOfBlobs = BF.BlobFinder(im1, tau)
            LstOfBlobs2 = BF.BlobFinder(im2, tau)
            
            #Use getbeads to pick out the objects that have a mass above a given number.
            LstOfBeads = BF.getbeads(P,LstOfBlobs)
            LstOfBeads2 = BF.getbeads(P,LstOfBlobs2)
            #Now check the distance from all beads in pic1 to all beads in pic2
            for bead in LstOfBeads:
                temp = []
                for bead2 in LstOfBeads2:
                    distance = bead.distanceTo(bead2)
                    #only append distances less than deltaP and distances not equal to zero.
                    #If a distance of zero is appended, the next if loop will keep appending
                    #zero instead of the minimum float value. 
                    if distance < deltaP and distance != 0.0:
                        temp.append(distance)
                    else:
                        continue
                #Append to distancesLst, then break out of this for loop and move on to the next one.
                if len(temp) != 0:
                    distancesLst.append(min(temp))
                else: continue

    #The functions in the following for loops are the same as above, the only differnce are the
    #pictures that are used.
    
    #run4/frame00010 through run4/frame00099
    for i in range(10,100):
        if i < 99:
            a = "run4/frame000" + str(i) + ".jpg"
            b = "run4/frame000"+ str(i+1) + ".jpg"
            im1 = Image.open(a)
            im2 = Image.open(b)

            LstOfBlobs = BF.BlobFinder(im1, tau)
            LstOfBlobs2 = BF.BlobFinder(im2, tau)

            LstOfBeads = BF.getbeads(P,LstOfBlobs)
            LstOfBeads2 = BF.getbeads(P,LstOfBlobs2)

            for bead in LstOfBeads:
                temp = []
                for bead2 in LstOfBeads2:
                    distance = bead.distanceTo(bead2)
                    if distance < deltaP and distance != 0.0:
                        temp.append(distance)
                    else:
                        continue
                if len(temp) != 0:
                    distancesLst.append(min(temp))
                else: continue
               
    #run4/frame00100 through run4/frame00199                    
    for i in range(100, 200):
        if i < 199:
            a = "run4/frame00" + str(i) + ".jpg"
            b = "run4/frame00"+ str(i+1) + ".jpg"
            im1 = Image.open(a)
            im2 = Image.open(b)

            LstOfBlobs = BF.BlobFinder(im1, tau)
            LstOfBlobs2 = BF.BlobFinder(im2, tau)

            LstOfBeads = BF.getbeads(P,LstOfBlobs)
            LstOfBeads2 = BF.getbeads(P,LstOfBlobs2)

            for bead in LstOfBeads:
                temp = []
                for bead2 in LstOfBeads2:
                    distance = bead.distanceTo(bead2)
                    if distance < deltaP and distance != 0.0:
                        temp.append(distance)
                    else:
                        continue
                if len(temp) != 0:
                    distancesLst.append(min(temp))
                else: continue
    
    #return text file.
    return np.savetxt('distances.txt', distancesLst, fmt='%.4f', header='List Of Distances')
```

```python
#call tracking function, no need to assign to a variable because it already does the savetxt step.
#Using mass size of 25 pixels, tau value of 600, and maximum distance change of 25 pixels.
TrackParticals(25, 600., 25) 
#load up text file.
np.loadtxt('distances.txt')
#see what it looks like
!cat 'distances.txt'
```

```python
def AvogadrosNumberEstimate(distancefile):
    """
    This function converts all the pixel distances to meters, uses this to find the 
    'self diffusion constant' D, estimates the Boltzmann constant k using the
    Stokes-Einstein relation, then calculates Avogadros number by dividing the gas
    constant by k.
    """
    r = 0 #initialize r
    conversion = 0.175e-6 #meters/pixel
    eta = 9.125e-4 #viscosity of water, N*s/(m^2)
    rho = 0.5e-6 #radius of bead, m
    R = 8.31457 #Gas constant, J/(K*mol)
    T = 297 #Room temperature, K
    
    #find the sum of all converted distances from the distance 
    #file squared to put through equation given.
    for dist in np.loadtxt(distancefile):
        r += (dist*conversion)**2

    #self diffusion constant estimation:
    D = r/(2*len(np.loadtxt(distancefile))) 
    
    #Boltzmann constant estimation:
    k = (6*D*np.pi*eta*rho)/T #m^2kg/Ks^2
    
    #Avogadros number estimation:
    Na = R/k 
    return Na 
```

```python
print "My Avogadros Number Estimate is: %.4e" %(AvogadrosNumberEstimate('distances.txt'))
print "Percent error = %.4f" %(((AvogadrosNumberEstimate('distances.txt')-6.0221413e23)/6.0221413e23)*100)
```
