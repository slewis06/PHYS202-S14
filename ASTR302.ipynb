```python
%pylab inline
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
labels = ['Blue', 'Yellow/White', 'Other']
fracs = [28, 98, 2]
plt.pie(fracs, labels=labels, colors=('b','y','r'), autopct='%1.1f%%', shadow=True)
plt.title('Elliptical Galaxies Color Distribution', fontsize=14)
plt.show()

print "Number of Ellipticals sampled: %.f \n Yellow/White: %.f \n Blue: %.f \n Other: %.f" \
    %(fracs[0]+fracs[1]+fracs[2], fracs[1], fracs[0], fracs[2])

print "\n"

plt.figure()

fracs2 = [25, 3, 1]
plt.pie(fracs2, labels=labels, colors=('b','y','r'), autopct='%1.1f%%', shadow=True)
plt.title('Spiral Galaxies Color Distribution', fontsize=14)
plt.show()

print "Number of Sprials sampled: %.f \n Yellow/White: %.f \n Blue: %.f \n Other: %.f" \
    %(fracs2[0]+fracs2[1]+fracs2[2], fracs2[1], fracs2[0], fracs2[2])

print "\n"
    
plt.figure()

fracs4 = [fracs[0] + fracs2[0],fracs[1] + fracs2[1],fracs[2] + fracs2[2]]
plt.pie(fracs4, labels=labels, colors=('b','y','r'), autopct='%1.1f%%', shadow=True)
plt.title('Total Distribution of Galaxy Color', fontsize=14)
plt.show()

print "Number of Galaxies sampled: %.f \n Yellow/White: %.f \n Blue: %.f \n Other: %.f" \
    %(fracs[0]+fracs[1]+fracs[2]+fracs2[0]+fracs2[1]+fracs2[2], fracs[1]+fracs2[1], \
      fracs[0]+fracs2[0], fracs[2]+fracs2[2])
print "\n"

plt.figure()

labels2 = ['Elliptical','Spiral','Irregular']
fracs3 = [fracs[0]+fracs[1]+fracs[2], fracs2[0]+fracs2[1]+fracs2[2],9]
plt.pie(fracs3, labels=labels2, autopct='%1.1f%%', shadow=True)
plt.title('Total Distribution of Galaxy Type', fontsize=14)
plt.show()

print "Number of Galaxies sampled: %.f \n Elliptical: %.f \n Spiral: %.f \n Irregular: %.f" \
    %(fracs[0]+fracs[1]+fracs[2]+fracs2[0]+fracs2[1]+fracs2[2], fracs[0]+fracs[1]+fracs[2], \
      fracs2[0]+fracs2[1]+fracs2[2], 9.0)
```
